```python
df['x_csdo_enddatetime'] = df['x_csdo_enddatetime'].replace('9999-12-30 23:59:59', '2100-12-30 23:59:59') 
#заменяем дату конца дейсвия на адекватную дату
```

```python
df['x_csdo_enddatetime'] = pd.to_datetime(df['x_csdo_enddatetime'], errors='coerce')  #заменим тип данных для вывода даты

# Вывод результата
print(df['x_csdo_enddatetime'])
```

```python
gr_1 = df.query('x_csdo_enddatetime > "2023-10-30 00:00:00"') 

gr_1
print("Количество актуальных записей по полю конечная дата:", gr_1.count(), "штук,")
```

```python
gr_2 = df.query('x_csdo_enddatetime < "2023-10-30 00:00:00"')
gr_2.count()
print("Количество не актуальных записей по полю конечная дата (меньше 30.10.2023 г.):", gr_2.count(), "штук,")
```

```python
remaining_df = df[~df['notification_number'].isin(gr_1['notification_number']) & ~df['notification_number'].isin(gr_2['notification_number'])]
print("Остаток (значения нулевые по полю Конечная дата):")
display(remaining_df)
```
